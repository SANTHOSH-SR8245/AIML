```python
def convert_price(deposit):
    if isinstance(deposit, str):
        deposit = deposit.replace(',', '')
        if 'L' in deposit:
            return float(deposit.replace('L', '').strip()) * 100000
        elif 'Cr' in deposit:
            return float(deposit.replace('Cr', '').strip()) * 10000000
        elif deposit.lower().strip() == 'no':
            return 0
    return float(deposit)

dataset['Security_deposit'] = dataset['Security_deposit'].apply(convert_price)
dataset['Rent_price'] = dataset['Rent_price'].apply(convert_price)
```
